```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from kerastuner.tuners import RandomSearch
```

```python
# Read data

data = pd.read_csv('musk_tweets_sent.csv')
```

```python
# Show first few rows

data.head()
```

```python
data.info()
```

```python
# Set date as the index

data.set_index('date', inplace = True)
```

```python
# Plot the data

data.plot()
```

```python
# Scale the data to values between 0 and 1

dataset = data['avg_sentiment_score'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range = (0, 1))
scaled_dataset = scaler.fit_transform(dataset)
```

```python
# Create train and test splits

train_size = int(len(scaled_dataset) * 0.8)
train, test = scaled_dataset[:train_size, :], scaled_dataset[train_size:, :]
```

```python
# Create a function which produces dataset with given look_back parameter

def create_dataset(dataset, look_back = 1):
  X, Y = [], []
  for i in range(len(dataset) - look_back - 1):
    X.append(dataset[i:(i + look_back), 0])
    Y.append(dataset[i + look_back, 0])
  return np.array(X), np.array(Y)
```

```python
# Create train and test splits in required format

look_back = 3
X_train, y_train = create_dataset(train, look_back)
X_test, y_test = create_dataset(test, look_back)
```

```python
# Convert data to requried shape

X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))
```

```python
# Define a function to build an LSTM model while defining a hyperparameter space to search

def build_model(hp):
  model = Sequential()
  model.add(LSTM(units = hp.Int('units', min_value = 30, max_value = 100, step = 10), input_shape = (1, look_back)))
  model.add(Dense(1))
  model.compile(loss = 'mean_squared_error', optimizer = hp.Choice('optimizer', values = ['adam', 'rmsprop']))
  return model
```

```python
# Initialize a RandomSearch object

tuner = RandomSearch(
    build_model,
    objective = 'val_loss',
    max_trials = 20,
    executions_per_trial = 3,
    directory = 'output',
    project_name = 'sentiment_lstm'
)
```

```python
# Search the hyperparameter space

tuner.search(X_train, y_train.reshape(-1, 1), epochs = 50, validation_split = 0.2, batch_size = 1, verbose = 1)
```

```python
# Get the best model and build it

best_hp = tuner.get_best_hyperparameters(num_trials = 1)[0]

model = build_model(best_hp)
```

```python
# Train the best model

history = model.fit(X_train, y_train.reshape(-1, 1), epochs = 50, validation_split = 0.2, batch_size = 1, verbose = 1)
```

```python
# Calculate test loss

test_loss = model.evaluate(X_test, y_test.reshape(-1, 1), verbose = 1)
print("Test loss:", test_loss)
```

```python
# Save the model

model.save('optimal_lstm')
```

```python
# Calculate train and test predictions

train_predict = model.predict(X_train)
test_predict = model.predict(X_test)
```

```python
# Perform inverse transformation

train_predict = scaler.inverse_transform(train_predict)
y_train = scaler.inverse_transform([y_train])
test_predict = scaler.inverse_transform(test_predict)
y_test = scaler.inverse_transform([y_test])
```

```python
# Calculate RMSE on train and test set

train_score = np.sqrt(mean_squared_error(y_train[0], train_predict[:, 0]))
print(f'Train Score: {train_score:.2f} RMSE')
test_score = np.sqrt(mean_squared_error(y_test[0], test_predict[:, 0]))
print(f'Test Score: {test_score:.2f} RMSE')
```

```python
# Plot Actual vs predicted on train and test set

train_predict_plot = np.empty_like(scaled_dataset)
train_predict_plot[:, :] = np.nan
train_predict_plot[look_back:len(train_predict) + look_back, :] = train_predict

test_predict_plot = np.empty_like(scaled_dataset)
test_predict_plot[:, :] = np.nan
test_predict_plot[len(train_predict) + (look_back * 2) + 1:len(scaled_dataset) - 1, :] = test_predict


actual_dataset = scaler.inverse_transform(scaled_dataset)

plt.figure(figsize = (12, 6))
plt.plot(actual_dataset, label = 'Actual Data')
plt.plot(train_predict_plot, label = 'Train Predictions')
plt.plot(test_predict_plot, label = 'Test Predictions')
plt.xlabel('Time')
plt.ylabel('Sentiment Score')
plt.title('Actual vs Predicted Data')
plt.legend()
plt.show()
```
